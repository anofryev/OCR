```python
def get_descriptors(img):   
        orb = cv2.ORB_create(nfeatures = 500, #500
                            scaleFactor = 1,#1.2
                            nlevels = 4,
                            edgeThreshold = 10,
                            firstLevel = 0,
                            WTA_K = 1,
                            patchSize = 10,
                            fastThreshold = 20)
        kp = orb.detect(img,None)
        kp, des = orb.compute(img, kp)

        return kp, des
```
